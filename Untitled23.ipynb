```python
import zipfile
import os


with zipfile.ZipFile("US Airline Flight Routes and Fares 1993-2024.csv.zip", 'r') as zip_ref:
    zip_ref.extractall("us_flight_data")

os.listdir("us_flight_data")
```

```python
file_path = "us_flight_data/US Airline Flight Routes and Fares 1993-2024.csv"
df = pd.read_csv(file_path)
```

```python
columns_to_drop = ['Geocoded_City1', 'Geocoded_City2', 'tbl', 'tbl1apk']
df.drop(columns=columns_to_drop, inplace=True, errors='ignore')
df.drop_duplicates(inplace=True)
```

```python
print("\nMissing Values per Column:\n", df.isnull().sum())
```

```python
df.dropna(inplace=True)
```

```python
df['route'] = df['city1'] + " ➔ " + df['city2']
```

```python
top_routes = df['route'].value_counts().head(10)
sns.barplot(x=top_routes.values, y=top_routes.index, palette="viridis")
plt.title("Top 10 Most Frequent Flight Routes")
plt.xlabel("Number of Records")
plt.ylabel("Route")
plt.tight_layout()
plt.show()
```

```python
avg_fare_quarter = df.groupby('quarter')['fare'].mean().reset_index()
sns.lineplot(data=avg_fare_quarter, x='quarter', y='fare', marker='o')
plt.title("Average Fare by Quarter")
plt.xlabel("Quarter")
plt.ylabel("Average Fare ($)")
plt.xticks([1, 2, 3, 4])
plt.tight_layout()
plt.show()
```

```python
correlation = df[['fare', 'nsmiles', 'passengers', 'fare_lg', 'fare_low']].corr()
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Numerical Features")
plt.tight_layout()
plt.show()
```

```python
sns.histplot(df['fare'], bins=50, kde=True, color='skyblue')
plt.title("Distribution of Flight Fares")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(x='nsmiles', y='passengers', data=df, alpha=0.3)
plt.title("Passenger Count vs Flight Distance")
plt.xlabel("Distance (Miles)")
plt.ylabel("Number of Passengers")
plt.tight_layout()
plt.show()
```
